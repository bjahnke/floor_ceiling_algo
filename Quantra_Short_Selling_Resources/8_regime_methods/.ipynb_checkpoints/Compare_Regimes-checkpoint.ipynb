# Comparison: Regime Change Detection Methods
In this notebook, we will compare all the three regime methods:
Method 1: Breakout,
Method 2: Returns,
and Method 3: Drawdowns


## Import libraries

```python
# Data manipulation libraries
import numpy as np
import pandas as pd
from datetime import datetime

# Calculates swing highs and lows
from scipy.signal import *
from scipy.signal import argrelextrema

# Appends new file paths to import modules
import sys
sys.path.append("..")

# All functions library
from data_modules.short_selling import swings
from data_modules.short_selling import regime_breakout_breakdown
from data_modules.short_selling import regime_sma
from data_modules.short_selling import regime_fc
from data_modules.short_selling import returns
from data_modules.short_selling import cumulative_returns

# Imports charting libraries
import matplotlib.pyplot as plt

%matplotlib inline
```

## Import data

### Files name:
1. Jan_2010_to_Jan_2019_Rebased_Series.csv, contains the rebased series for 'Bank of America'

This data is available in the last unit of this course. You can find this file in the downloadables section of the course.

This data is available in the last unit of this course. You can find this file in the downloadables section of the course. We import these data by using the pandas function 'pandas.read_csv()'. This function will read the comma-separated values (CSV) file into dataframe.


```python
# Data is stored in the directory 'data_modules' 
relative_path = "../data_modules/"
# Reads and imports the rebased series using the function 'read_csv'
data = pd.read_csv(
    relative_path + 'Jan_2010_to_Jan_2019_Rebased_Series.csv', index_col=0, parse_dates=True)

# Set index in datetime format
data.index = pd.to_datetime(data.index)

# Returns the first two rows of the dataset
data.head(2)
```

## Method 1: Breakout-Breakdown

In this method, we use relative series to calculate

1. Regime breakout
2. Regime simple moving average
3. Swing highs and lows
4. Regime floor and ceiling
6. and, plot the regime change

```python
# 1. Calculate regime breakout using the 'regime_breakout' for a breakout period of 252 days
data = regime_breakout_breakdown(
    breakout_period=252, df=data, high='rebased_high', low='rebased_low')

# 2. Calculate the regime simple moving average using 'regime_sma' function
data = regime_sma(df=data, price='rebased_close', short_term=20, long_term=200)

# 3. Calculate swings using the 'swings' function
data = swings(df=data, high='rebased_high',
              low='rebased_low', argrel_window=20)

# 4. Using the 'regime_fc' function, calculate the floor and ceiling
data = regime_fc(data, 'rebased_close', 'srebased_low',
                 'srebased_high', threshold=1.5, t_dev=63, decimals=2)

# Returns the bottom 2 rows of the dataset
data.tail(2)
```

```python
# 5. Plot the regime change
# Here, 'rg' stands for regime and 'r' for relative, refer to the code to understand the label
data[['rebased_close', 'r_regime_floorceiling', 'regime_breakout_breakdown_252', 'regime_sma_20_200', 'r_floor', 'r_ceiling']].plot(
    style=['k', 'mo', 'c', 'b', 'g^', 'rv'],
    secondary_y=['rebased_close', 'r_floor', 'r_ceiling', 'srebased_high', 'srebased_low'], figsize=(10, 7))

# Add a legend
plt.legend()
# Display the plot
plt.title('Breakout-Breakdown', fontsize=16)
plt.ylabel('Price', fontsize=14)
plt.xlabel('Date', fontsize=14)
# Add a legend
plt.legend()
# Add a grid
plt.grid()
# Display the graph
plt.show()
```

## Method 2: Returns


In part 2, we will calculate cumulative returns and plot them along with the passive relative returns.
Calculate relative and cumulative relative returns for
1. Rebased close price
2. Regime breakout
3. Moving average 
4. Regime floorceiling 
5. And, plot all these cumulative returns

The Floor/Ceiling method has built-in hindsight bias which means swings are discovered and retroactively assigned. We, therefore, need to introduce a lag, which can be lesser than or equal to the argel window.

```python
# Declare argel_window
argrel_window = 20
# Declare argel_lag
argrel_lag = 20 
```

```python
# 1. Calculate relative and cumulative relative returns for rebased close price
data['relative_returns'] = returns(data['rebased_close'])
data['cum_relative_returns'] = cumulative_returns(data['relative_returns'])

# 2. Calculate relative and cumulative relative returns for regime breakout using the relative returns
data['regime_breakout_breakdown_252_returns'] = data['relative_returns'] * \
    data['regime_breakout_breakdown_252'].shift(1)
data['cum_regime_breakout_breakdown_252_returns'] = cumulative_returns(
    data['regime_breakout_breakdown_252_returns'])

# 3. Calculate relative and cumulative relative returns for moving average using the relative returns
data['regime_sma_returns'] = data['relative_returns'] * \
    data['regime_sma_20_200'].shift(1)
data['cum_regime_sma_20_200_returns'] = cumulative_returns(
    data['regime_sma_returns'])

# 4. Calculate relative and cumulative relative returns for regime floorceiling using the relative returns
data['floorceiling_returns'] = data['relative_returns'] * \
    data['r_regime_floorceiling'].shift(argrel_lag)
data['cum_regime_floorceiling_returns'] = cumulative_returns(
    data['floorceiling_returns'])

# Return the bottom 2 rows of the dataset
data.tail(2)
```

```python
# 5. And, plot all these cumulative returns
data[['cum_relative_returns', 'cum_regime_floorceiling_returns', 'cum_regime_breakout_breakdown_252_returns', 'cum_regime_sma_20_200_returns']].plot(
    style=['k', 'm', 'c', 'b'], figsize=(10, 7))
plt.title('Cumulative Returns', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Returns', fontsize=14)

# Add a legend
plt.legend()
# Add a grid
plt.grid()
# Display the graph
plt.show()
```

## Method 3: Drawdowns
A drawdown refers to how much an investment or trading account is down from the peak before it recovers back to the peak. Drawdowns are the difference between cumulative and absolute returns.

In this method, we calculate drawdown for
1. Relative returns
2. Regime breakout
3. Regime simple moving average
4. Regime floor and ceiling
5. Plot the drawdowns

```python
# 1. Calculate drawdown for relative returns
data['drawdown'] = data['cum_relative_returns'].cummax() - \
    np.abs(data['cum_relative_returns'])

# 2. Calculate drawdown for regime breakout
data['breakout_drawdown'] = data['cum_regime_breakout_breakdown_252_returns'].cummax(
) - np.abs(data['cum_regime_breakout_breakdown_252_returns'])

# 3. Calculate drawdown for regime simple moving average
data['sma_20_200_drawdown'] = data['cum_regime_sma_20_200_returns'].cummax() - \
    np.abs(data['cum_regime_sma_20_200_returns'])

# 4. Calculate drawdown for regime floor and ceiling
data['floorceiling_drawdown'] = data['cum_regime_floorceiling_returns'].cummax(
) - np.abs(data['cum_regime_floorceiling_returns'])

# Returns the bottom 2 rows of the dataset
data.tail(2)
```

```python
# 5. Plot drawdowns
data[['drawdown', 'floorceiling_drawdown', 'breakout_drawdown', 'sma_20_200_drawdown']].plot(style=['k:', 'm:', 'c:', 'b:'],
                                                                                             figsize=(10, 7))
plt.title('Drawdown', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Returns', fontsize=14)

# Add a legend
plt.legend()
# Add a grid
plt.grid()
# Display the graph
plt.show()
```
